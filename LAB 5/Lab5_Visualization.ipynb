# **1.IMPORTING DATASET AND** **LIBRARIES**

```python
import pandas as pd

df = pd.read_csv("Covid Data.csv")
df.head()
```

# **2.PRINTING COLUMNS**

```python
print(df.columns)
```

# **3.Histogram (AGE)**

```python
import matplotlib.pyplot as plt

plt.hist(df["AGE"])
plt.title("Age Distribution")
plt.show()
```

# **4.BAR CHART**

```python
df["SEX"].value_counts().plot(kind="bar")
plt.title("Gender Distribution")
plt.show()
```

# 5.**Scatter** **plot**

```python
plt.scatter(df["AGE"], df["PNEUMONIA"])
plt.title("Age vs Pneumonia")
plt.show()
```

# **6.BOX PLOT**

```python
import seaborn as sns

sns.boxplot(x=df["AGE"])
plt.show()
```

# 7.**VIOLIN** **PLOT**

```python
sns.violinplot(x=df["AGE"])
plt.show()
```

# 8.**HEATMAP** **PLOT**

```python
import seaborn as sns

# Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])

corr = numeric_df.corr()

sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

# **9.CUSTOMISED PLOT**

```python
df["AGE"].plot(kind="hist", bins=20)
plt.title("Customized Age Histogram")
plt.xlabel("Age")
plt.show()
```

# **9. SUBPLOT**

```python
plt.subplot(1,2,1)
plt.hist(df["AGE"])

plt.subplot(1,2,2)
df["SEX"].value_counts().plot(kind="bar")

plt.show()
```

# **10 PIE PLOT**

```python
df["SEX"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.show()
```

# **11. Extra Chart (ICU vs Age)**

```python
plt.scatter(df["AGE"], df["ICU"])
plt.show()
```

# **12. LINE PLOT**

```python
df["AGE"].head(50).plot()
plt.title("Sample Age Trend")
plt.show()
```
