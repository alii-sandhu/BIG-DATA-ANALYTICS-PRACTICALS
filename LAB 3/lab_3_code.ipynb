# **1.import numpy**

```python
import numpy as np
```

# **2.manuplating numpy arrays**

```python
arr1 = np.array([1,2,3,4])
arr2 = np.array([[1,2,3],[4,5,6]])
arr3 = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])

print(arr1)
print(arr2)
print(arr3)
```

# **3.Indexing and slicing**

```python
arr = np.array([10,20,30,40,50])
print(arr[0])
print(arr[1:4])
```

# **4**.Reshaping

```python
arr = np.array([1,2,3,4,5,6])
print(arr.reshape(2,3))
```

# **5**.**Math operations**

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

print(a+b)
print(a*b)
print(a+5)
```

# **6**.**Statistics**

```python
data = np.array([10,20,30,40,50])

print(np.mean(data))
print(np.median(data))
print(np.std(data))
print(np.percentile(data,25))
```

# **7**.**Linear Algebra**

```python
x = np.array([1,2])
y = np.array([3,4])

print(np.dot(x,y))

A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

print(np.matmul(A,B))
```

# **8.Random Numbers**

```python
import time

lst = list(range(1000000))

start = time.time()
sum(lst)
print("List:", time.time()-start)

arr = np.array(lst)

start = time.time()
np.sum(arr)
print("NumPy:", time.time()-start)
```
