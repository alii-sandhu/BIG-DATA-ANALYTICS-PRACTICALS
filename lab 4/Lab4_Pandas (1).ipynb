# **1.IMPORTING LIBRARIES**

```python
import pandas as pd
import numpy as np
```

# **2.LOADING DATASET**

```python
df = pd.read_csv("/tested.csv")
```

# **3.EXPLORING DATA**

```python
print(df.head())
print(df.tail())
print(df.info())
print(df.describe())
print(df.shape)
```

# **4.SELECTING COLUMNS**

```python
print(df["Name"])
print(df[["Name", "Age"]])
```

# **5. FILTERING ROWS**

```python
# Age > 23
print(df[df["Age"] > 23])

# Survived passengers
print(df[df["Survived"] == 1])
```

# **6.SORTING DATA**

```python
# Single column
print(df.sort_values(by="Age"))

# Multiple columns
print(df.sort_values(by=["Age", "Fare"]))
```

# **7.HANDLING MISSING DATA**

```python
df.loc[2, "Age"] = np.nan
df.loc[4, "Fare"] = np.nan
```

```python
print(df.isnull())
```

```python
print(df.dropna())
```

```python
print(df.fillna(0))
```

# **8.COLUMN OPERATIONS**

```python
# Add column
df["Discount"] = df["Fare"] * 0.1

# Remove column
df = df.drop("Discount", axis=1)

# Rename column
df = df.rename(columns={"Fare": "TicketPrice"})

print(df)
```

```python
df.to_csv("output.csv", index=False)
```
